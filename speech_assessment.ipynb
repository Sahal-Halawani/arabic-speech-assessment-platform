````python
prompt = f"""
You are a professional Arabic language examiner.

Evaluate the following Arabic speaking response for an Arabic learning platform entry exam.

Assessment Criteria:
1. Grammar
2. Fluency
3. Vocabulary
4. Coherence

Scoring Rules:
- Scores must be integers from 1 to 10.
- Use realistic and strict scoring.

Level Rules:
- Use ONLY one CEFR level:
A1, A2, B1, B2, C1, or C2

Feedback Rules:
- Feedback MUST be in English only.
- Maximum 2 short sentences.

Return ONLY valid JSON in this exact format:

{{
  "Grammar": 0,
  "Fluency": 0,
  "Vocabulary": 0,
  "Coherence": 0,
  "Level": "B1",
  "Feedback": "Short professional feedback."
}}

Do not add explanations.
Do not add markdown.
Do not add extra text.

Arabic Text:
{text_small}
"""
````

````python
# LLM 1
````

````python
import google.generativeai as genai

genai.configure(api_key="GEMINI_API_KEY")

gemini = genai.GenerativeModel("models/gemini-2.5-flash")

gemini_response = gemini.generate_content(prompt)

gemini_output = gemini_response.text

print("=== Gemini Result ===")
print(gemini_output)
````

````python
# LLM 2
````

````python
from openai import OpenAI

groq_client = OpenAI(
    api_key="GROQ_API_KEY",
    base_url="https://api.groq.com/openai/v1"
)

llama_response = groq_client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {"role": "user", "content": prompt}
    ]
)

llama_output = llama_response.choices[0].message.content

print("=== LLaMA 3.3 Result ===")
print(llama_output)
````

````python
# LLM 3
````

````python
from huggingface_hub import InferenceClient

client = InferenceClient(
    api_key="HF_TOKEN"
)

response = client.chat_completion(
    model="Qwen/Qwen2.5-72B-Instruct",
    messages=[
        {
            "role": "user",
            "content": prompt
        }
    ],
    max_tokens=200
)

hf_output = response.choices[0].message.content

print("=== HuggingFace Result ===")
print(hf_output)
````

````python
# لترتيب الجدول
````

````python
import json

gemini_scores = json.loads(gemini_output)
llama_scores = json.loads(llama_output)
hf_scores = json.loads(hf_output)
````

````python
%%writefile app.py

import streamlit as st
import whisper
import json
import pandas as pd
import google.generativeai as genai
from openai import OpenAI
from huggingface_hub import InferenceClient
import tempfile
import re

# =====================================
# Page Config
# =====================================

st.set_page_config(
    page_title="Arabic Speech Assessment",
    page_icon="🎤",
    layout="wide"
)

# =====================================
# Session State
# =====================================

if "authenticated" not in st.session_state:
    st.session_state.authenticated = False

# =====================================
# Helper Functions
# =====================================

def clean_json_response(text):

    text = text.strip()

    text = re.sub(r"```json", "", text)
    text = re.sub(r"```", "", text)

    return text.strip()

# =====================================
# API PAGE
# =====================================

if not st.session_state.authenticated:

    st.title("🔑 API Configuration")

    st.write(
        "Enter your API keys to continue."
    )

    st.divider()

    gemini_key = st.text_input(
        "Gemini API Key",
        type="password"
    )

    groq_key = st.text_input(
        "Groq API Key",
        type="password"
    )

    hf_key = st.text_input(
        "HuggingFace API Key",
        type="password"
    )

    st.divider()

    if st.button("Continue to Exam"):

        if (
            gemini_key == ""
            or groq_key == ""
            or hf_key == ""
        ):

            st.warning(
                "Please enter all API keys."
            )

        else:

            st.session_state.gemini_key = gemini_key
            st.session_state.groq_key = groq_key
            st.session_state.hf_key = hf_key

            st.session_state.authenticated = True

            st.rerun()

# =====================================
# MAIN APP
# =====================================

else:

    st.title("🎤 Arabic Speech Assessment Platform")

    st.write(
        "Automated Arabic Entry Exam using Whisper + Gemini + LLaMA + Qwen"
    )

    st.divider()

    # =====================================
    # Sidebar
    # =====================================

    with st.sidebar:

        st.header("⚙️ Settings")

        whisper_size = st.selectbox(
            "Whisper Model Size",
            ["base", "small", "medium"],
            index=1
        )

        st.info(
            "Medium model gives better Arabic accuracy."
        )

        if st.button("Logout"):

            st.session_state.authenticated = False
            st.rerun()

    # =====================================
    # Question
    # =====================================

    st.subheader("Arabic Entry Exam Question")

    question = """
Introduce yourself in Arabic.
Talk about your studies, interests, and future goals.
"""

    st.info(question)

    st.divider()

    # =====================================
    # Audio Input
    # =====================================

    st.subheader("Upload or Record Your Answer")

    uploaded_audio = st.file_uploader(
        "Upload Arabic Audio File",
        type=["mp3", "wav", "m4a"]
    )

    recorded_audio = st.audio_input(
        "Or record your voice directly"
    )

    st.divider()

    # =====================================
    # Start Evaluation
    # =====================================

    if st.button("Start Evaluation"):

        audio_source = None

        if uploaded_audio is not None:
            audio_source = uploaded_audio

        elif recorded_audio is not None:
            audio_source = recorded_audio

        if audio_source is None:

            st.warning(
                "Please upload or record audio."
            )

        else:

            # =====================================
            # Save Audio
            # =====================================

            audio_bytes = audio_source.read()

            with tempfile.NamedTemporaryFile(
                delete=False,
                suffix=".wav"
            ) as tmp:

                tmp.write(audio_bytes)

                audio_path = tmp.name

            st.success(
                "Audio received successfully."
            )

            st.audio(
                audio_bytes,
                format="audio/wav"
            )

            st.divider()

            # =====================================
            # Whisper
            # =====================================

            with st.spinner(
                "Transcribing Arabic speech..."
            ):

                whisper_model = whisper.load_model(
                    whisper_size
                )

                result = whisper_model.transcribe(
                    audio_path,
                    language="ar",
                    fp16=False
                )

                text_small = result["text"]

            st.subheader(
                "📝 Transcribed Arabic Text"
            )

            st.write(text_small)

            st.divider()

            # =====================================
            # Prompt
            # =====================================

            prompt = f'''
You are a professional Arabic language examiner.

Evaluate the following Arabic speaking response for an Arabic learning platform entry exam.

Assessment Criteria:
1. Grammar
2. Fluency
3. Vocabulary
4. Coherence

Scoring Rules:
- Scores must be integers from 1 to 10.
- Use realistic and strict scoring.
- Be highly accurate and fair.

Level Rules:
- Use ONLY one CEFR level:
A1, A2, B1, B2, C1, or C2

Feedback Rules:
- Feedback MUST be in English only.
- Maximum 2 short sentences.
- Mention one strength and one weakness.

Return ONLY valid JSON in this exact format:

{{
  "Grammar": 0,
  "Fluency": 0,
  "Vocabulary": 0,
  "Coherence": 0,
  "Level": "B1",
  "Feedback": "Short professional feedback."
}}

Do not add explanations.
Do not add markdown.
Do not add extra text.

Arabic Text:
{text_small}
'''

            # =====================================
            # Gemini
            # =====================================

            with st.spinner(
                "Evaluating using Gemini..."
            ):

                genai.configure(
                    api_key=st.session_state.gemini_key
                )

                gemini = genai.GenerativeModel(
                    "models/gemini-2.5-flash"
                )

                gemini_response = gemini.generate_content(
                    prompt,
                    generation_config={
                        "temperature": 0
                    }
                )

                gemini_output = json.loads(
                    clean_json_response(
                        gemini_response.text
                    )
                )

            # =====================================
            # Groq Client
            # =====================================

            groq_client = OpenAI(
                api_key=st.session_state.groq_key,
                base_url="https://api.groq.com/openai/v1"
            )

            # =====================================
            # LLaMA 70B
            # =====================================

            with st.spinner(
                "Evaluating using LLaMA 70B..."
            ):

                llama_response = groq_client.chat.completions.create(
                    model="llama-3.3-70b-versatile",
                    temperature=0,
                    messages=[
                        {
                            "role": "user",
                            "content": prompt
                        }
                    ]
                )

                llama_output = json.loads(
                    clean_json_response(
                        llama_response.choices[0].message.content
                    )
                )

            # =====================================
            # HuggingFace Qwen
            # =====================================

            with st.spinner(
                "Evaluating using HuggingFace Qwen..."
            ):

                hf_client = InferenceClient(
                    api_key=st.session_state.hf_key
                )

                hf_response = hf_client.chat_completion(
                    model="Qwen/Qwen2.5-72B-Instruct",
                    messages=[
                        {
                            "role": "user",
                            "content": prompt
                        }
                    ],
                    max_tokens=200
                )

                hf_output = json.loads(
                    clean_json_response(
                        hf_response.choices[0].message.content
                    )
                )

            st.divider()

            # =====================================
            # DataFrame
            # =====================================

            df = pd.DataFrame([
                {"Model": "Gemini", **gemini_output},
                {"Model": "LLaMA 70B", **llama_output},
                {"Model": "HuggingFace Qwen", **hf_output}
            ])

            st.subheader(
                "📊 LLM Comparison Results"
            )

            st.dataframe(
                df,
                use_container_width=True
            )

            st.divider()

            # =====================================
            # Average Scores
            # =====================================

            numeric_cols = [
                "Grammar",
                "Fluency",
                "Vocabulary",
                "Coherence"
            ]

            df["Average"] = df[
                numeric_cols
            ].mean(axis=1)

            st.subheader(
                "⭐ Average Scores"
            )

            st.dataframe(
                df[["Model", "Average"]],
                use_container_width=True
            )

            # =====================================
            # Best Model
            # =====================================

            best_model = df.loc[
                df["Average"].idxmax()
            ]

            st.success(
                f"Best Performing Model: "
                f"{best_model['Model']} "
                f"with average score "
                f"{best_model['Average']:.2f}"
            )

            st.divider()

            # =====================================
            # Chart
            # =====================================

            st.subheader(
                "📈 Comparison of LLM Scores"
            )

            chart_df = df.set_index(
                "Model"
            )[numeric_cols]

            st.bar_chart(chart_df)

            st.divider()

            # =====================================
            # Feedback
            # =====================================

            st.subheader(
                "💬 Individual Model Feedback"
            )

            for _, row in df.iterrows():

                with st.expander(
                    f"{row['Model']} Feedback"
                ):

                    st.write(
                        f"**Level:** {row['Level']}"
                    )

                    st.write(
                        f"**Feedback:** {row['Feedback']}"
                    )
````
